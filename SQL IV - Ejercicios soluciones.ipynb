```python
query = """
    SELECT
id, 
TRIM(MAX(name)) as name,
year,
CAST(CASE WHEN AVG(ENE) IS NULL THEN 0 ELSE AVG(ENE) END AS float) as ENE,
CAST(CASE WHEN AVG(FEB) IS NULL THEN 0 ELSE AVG(FEB) END AS float) as FEB,
CAST(CASE WHEN AVG(MAR) IS NULL THEN 0 ELSE AVG(MAR) END AS float) as MAR,
CAST(CASE WHEN AVG(ABR) IS NULL THEN 0 ELSE AVG(ABR) END AS float) as ABR,
CAST(CASE WHEN AVG(MAY) IS NULL THEN 0 ELSE AVG(MAY) END AS float) as MAY,
CAST(CASE WHEN AVG(JUN) IS NULL THEN 0 ELSE AVG(JUN) END AS float) as JUN,
CAST(CASE WHEN AVG(JUL) IS NULL THEN 0 ELSE AVG(JUL) END AS float) as JUL,
CAST(CASE WHEN AVG(AGO) IS NULL THEN 0 ELSE AVG(AGO) END AS float) as AGO,
CAST(CASE WHEN AVG(SEP) IS NULL THEN 0 ELSE AVG(SEP) END AS float) as SEP,
CAST(CASE WHEN AVG(OCT) IS NULL THEN 0 ELSE AVG(OCT) END AS float) as OCT,
CAST(CASE WHEN AVG(NOV) IS NULL THEN 0 ELSE AVG(NOV) END AS float) as NOV,
CAST(CASE WHEN AVG(DIC) IS NULL THEN 0 ELSE AVG(DIC) END AS float) as DIC
FROM (
	SELECT
		pre.id,
		pre.name,
		pre.year,
		CASE
			WHEN pre.mes = 1 THEN precio
		END as ENE,
		CASE
			WHEN pre.mes = 2 THEN precio
		END as FEB,
		CASE
			WHEN pre.mes = 3 THEN precio
		END as MAR,
		CASE
			WHEN pre.mes = 4 THEN precio
		END as ABR,
		CASE
			WHEN pre.mes = 5 THEN precio
		END as MAY,
		CASE
			WHEN pre.mes = 6 THEN precio
		END as JUN,
		CASE
			WHEN pre.mes = 7 THEN precio
		END as JUL,
		CASE
			WHEN pre.mes = 8 THEN precio
		END as AGO,
		CASE
			WHEN pre.mes = 9 THEN precio
		END as SEP,
		CASE
			WHEN pre.mes = 10 THEN precio
		END as OCT,
		CASE
			WHEN pre.mes = 11 THEN precio
		END as NOV,
		CASE
			WHEN pre.mes = 12 THEN precio
		END as DIC
		FROM (
			SELECT
				base.id,
				base.name,
				base.year,
				base.mes,
				CAST(REPLACE(base.price, '$', '') as float) as precio
			FROM (
				SELECT 
					airbnb_listings.id,
					airbnb_listings.name,
					airbnb_calendar.*,
					YEAR(airbnb_calendar.date) as year, 
					MONTH(airbnb_calendar.date) as mes 
					FROM airbnb_calendar LEFT JOIN airbnb_listings ON 
					airbnb_listings.id = airbnb_calendar.listing_id
			) base
		) pre

) end

GROUP BY id, year
"""

sql(query, dw)
```
